```python
def  binarySearchIndexAndComparisons(list_to_sort : list , value_to_check) -> tuple:
    search_result = False
    comparison = 0 
    list_to_sort.sort()
    
    left , right = 0 , len(list_to_sort)-1 
    
    while left <= right and not search_result :
        mid = (left+right)//2 
        comparison += 1 
        if value_to_check == list_to_sort[mid]:
            search_result = True
        elif value_to_check < list_to_sort[mid]:
            right = mid-1 
        else:
            left = mid+1 
    return (search_result, comparison) 
    
```

```python
from collections import defaultdict
def combinationSort(strList: list) -> list:
    data_dict = defaultdict(list)
    first_list = []
    second_list = []
    
    for  value in strList:
        data_dict[value[0]].append(value)
    for key in sorted(data_dict.keys()):
        list_to_add = data_dict[key]
        if len(list_to_add) > 1 :
            sorted_list = sorted(list_to_add, reverse=True)
            first_list.extend(list_to_add)
            second_list.extend(sorted_list)
        else:
            first_list.extend(list_to_add)
            second_list.extend(list_to_add)
    return first_list, second_list
combinationSort(["d34", "g54", "d12", "b87", "g1", "c65", "g40", "g5", "d77"])
```

```python
def findLargest(L):
    left , right = 0 , len(L)-1
    if len(L) ==1:
        return L[0]
    largest = 0 
    while left < right :
        
        mid = (left+right)//2 
        
        largest = max(L[left] , L[right] , L[mid] , largest)
        if mid > L[left]:
            left = mid+1 
        else:
            right = mid -1 
    return largest 

```

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        from collections import Counter
        
        data_counter = Counter(nums)
        
        result = []
        
        for key , value in data_counter.items():
            if value >= k:
                result.append(key)
        return result 
```
