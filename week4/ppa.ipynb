```python


class Queue:
    def __init__(self) -> None:
        self.queue = []
        
    def Isempty(self):
        return self.queue == []
    
    def Equeue(self, value):
        self.queue.append(value)
    def Dqueue(self,):
        if not self.Isempty():
            return self.queue.pop(0)   
        else:
            return []
        
def findComponents_undirectedGraph(vertices , edges):
    
    # create a dictionary to keep track of visited 
    visited = {}
     # adjacency list 
    adj_list = {}
    for vertex in vertices:
        adj_list[vertex] = set()
        visited[vertex] = False
    
    for i , j in edges:
        adj_list[i].add(j)
        adj_list[j].add(i)
        
    my_queue = Queue()
    component = 0 
    for v in vertices:
        if not visited[v]:
            my_queue.Equeue(v)
            while not my_queue.Isempty():
                to_visit = my_queue.Dqueue()
                if not visited[to_visit]:
                    for neighbor in adj_list[to_visit]:
                        my_queue.Equeue(neighbor)
                    visited[to_visit] = True
            component+=1 
                
    
            
    return component
```

```python
def findAllPaths(v, Alist, source, dest):
    
    all_paths = []
    
    my_stack = [(source, [source])]  # initialize the source and path 
    
    while my_stack:
        
        current , path = my_stack.pop()
        
        if current == dest:
            all_paths.append(path)
        else:
            for neighbour in  Alist.get(current, []):
                if neighbour not in path:
                    my_stack.append((neighbour , path + [neighbour]))
    return all_paths
```

```python
from collections import deque

def bfs(maze, start, end):
    rows, cols = len(maze), max(len(row) for row in maze)
    queue = deque([start])
    visited = set([start])
    steps = 0

    while queue:
        for _ in range(len(queue)):
            x, y = queue.popleft()
            if (x, y) == end:
                return steps

            for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < rows and 0 <= ny < cols and (nx, ny) not in visited:
                    if ny < len(maze[nx]) and maze[nx][ny] != 'X':
                        queue.append((nx, ny))
                        visited.add((nx, ny))

        steps += 1

    return -1

def solve_maze(maze):
    rows, cols = len(maze), max(len(row) for row in maze)

    # Find start (left-most wall opening), key position, and exit (right-most wall opening)
    start = None
    key_pos = None
    exit_pos = None

    for r in range(rows):
        if len(maze[r]) > 0 and maze[r][0] == ' ':
            start = (r, 0)
        if len(maze[r]) > 0 and maze[r][-1] == ' ':
            exit_pos = (r, len(maze[r])-1)
        for c in range(len(maze[r])):
            if maze[r][c] == '*':
                key_pos = (r, c)
    
    # Step 1: Find shortest path from start to key
    if key_pos is None:
        return -1  # Key is not present in the maze
    
    steps_to_key = bfs(maze, start, key_pos)
    if steps_to_key == -1:
        return -1  # Key is not reachable

    # Step 2: Find shortest path from key to exit
    steps_to_exit = bfs(maze, key_pos, exit_pos)
    if steps_to_exit == -1:
        return -2  # Exit is not reachable after reaching the key

    return steps_to_key + steps_to_exit

# Input Parsing
input_lines = [
    "1XXXXXXXXXXXXXX",
    "2XXXXX",
    "3XXX XX",
    "4XX5XXXX XX",
    "6XXXX7X",
    "8XX9XX",
    "10XXXXXXXXXXXXXX"
]

# Extract the maze lines
maze = []
for line in input_lines:
    maze.append(line[1:].strip())  # Remove the first character (row number)

# Convert maze to list of lists for easier manipulation
maze = [list(row) for row in maze]
result = solve_maze(maze)
print(result)
```
