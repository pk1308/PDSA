```python
def fun(m,n):
    if m==n:
        return n 
    else:
        if m > n:
            return fun(m=m-n, n =n )
        else:
            return fun(m , n-m)
```

```python
fun(25,100)
```

```python
def fun(N):
    total  = 0 
    for i in range(N//2, N ):
        j = 2
        while (j <=N):
            total = total + N/2
            j = j * 2 
    return total 
```

```python
fun(1024)
```
